```python
import numpy as np
import matplotlib.pyplot as plt

def psi(r,theta):
    return ( (r**2)*np.exp(-r)*np.cos(theta) )

degrees = np.array([0,30,45,60,90])
degrees_r = degrees*(np.pi/180)

prob = []
radius_range = np.arange(0,4,0.1)
for bohr_radius in radius_range:
    prob.append(psi(bohr_radius,degrees_r))
prob = np.array(prob)
prob = prob.T

for i in range(len(degrees)):
    plt.plot(radius_range,prob[i], label = f"theta = {degrees[i]}")
plt.xlabel("Bohr Radius")
plt.ylabel("Probability")
plt.legend()
plt.show()
```
